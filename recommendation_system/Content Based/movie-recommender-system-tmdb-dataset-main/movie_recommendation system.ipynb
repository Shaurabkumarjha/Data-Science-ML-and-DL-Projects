```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies['original_language'].value_counts()
# Not consider language column beacuse 90 percent of data is in english language
```

```python
#geners
#keywords
#id
#title
#overview
#cast
#crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
    
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
def convert1(text):
    L=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter +=1
    return L
```

```python
# movies['cast']=movies['cast'].apply(convert1)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def removeSps(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['crew']=movies['crew'].apply(removeSps)
movies['cast']=movies['cast'].apply(removeSps)
movies['genres']=movies['genres'].apply(removeSps)
movies['keywords']=movies['keywords'].apply(removeSps)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
